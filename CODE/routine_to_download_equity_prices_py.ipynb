```python
ds = DSWS.Datastream(username = 'annalisa.molino@prometeia.com', password = 'Amif2024')

# Define batch size  
batch_size = 50  
tot_ISIN = len(my_ISINS)
#output_file = "G:/CLIMATE RISK & ESG/PROGETTI/Reputation_risk/REPUTATION/DATA/equity_1k_2012_2024.xlsx"
output_file = "G:/CLIMATE RISK & ESG/PROGETTI/Reputation_risk/REPUTATION/DATA/equity_3k_2012_2024.xlsx"
#output_file = "G:/CLIMATE RISK & ESG/PROGETTI/Reputation_risk/REP_RISK_ISP_2025/Code/DATA/equity_prices_ISP_W_2023_2025.xlsx"
#output_file = "G:/CLIMATE RISK & ESG/PROGETTI/Reputation_risk/REPUTATION/DATA//equity_prices_others_july_2025.xlsx"

# Create an Excel writer to store multiple sheets
with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
    for i in range(0, tot_ISIN, batch_size):  
        # Get 50 tickers at a time  
        tickers_str = ", ".join(my_ISINS["Modified_ISIN"].iloc[i:i + batch_size].tolist())  

        # Fetch data  
        try:
            response = ds.get_data(tickers=tickers_str,  
                                   start='2010-01-01', end='2025-05-30', freq="D")
            # Save each batch to a separate sheet  
            sheet_name = f"Batch_{i//batch_size + 1}"  # Sheet names: Batch_1, Batch_2, ...
            response.to_excel(writer, sheet_name=sheet_name, index=True)
            print(f"✅ Saved batch {i//batch_size + 1} to sheet: {sheet_name}")
        
        except Exception as e:
            print(f"❌ Error in batch {i//batch_size + 1}: {e}")

print(f"✅ All batches saved in {output_file}")

```
